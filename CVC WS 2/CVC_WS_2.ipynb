## Task 1

My function

```python
def RGB2HSV(img):
  lista = img.tolist()
  for row in range(len(lista)):
    for col in range(len(lista[row])):
      r = lista[row][col][0] / 255
      g = lista[row][col][1] / 255
      b = lista[row][col][2] / 255
      Cmax = max(r, g, b)
      Cmin = min(r, g, b)
      delta = Cmax - Cmin

      #Hue
      if delta == 0:
        h = 0
      elif Cmax == r:
        h = 60 * (((g-b)/delta) % 6)
      elif Cmax == g:
        h = 60 * (((b-r)/delta) + 2)
      elif Cmax == b:
        h = 60 * (((r-g)/delta) + 4)
      #Value
      v = Cmax
      #Saturation
      if Cmax == 0:
        s = 0
      else:
        s = delta / Cmax
      lista[row][col][0] = h
      lista[row][col][1] = s
      lista[row][col][2] = v
  img = np.array(lista)
  return io.imshow(img)
```

```python
def change_contrast_and_brightness(img, contrast_value, brightness_value):
    new_img = np.empty_like(img)
    for y in range(img.shape[0]):
      for x in range(img.shape[1]):
        for c in range(img.shape[2]):
          new_img[y,x,c] = np.clip(contrast_value * img[y,x,c] + brightness_value, 0, 255)
    return new_img
```

```python
img = color = np.array(
    [
     [[242, 245, 66],
      [242, 151, 65],
      [237, 62, 164]
      ],
     [
      [64, 237, 234],
      [96, 237, 64],
      [64, 78, 237]
      ],
     [
      [237, 64, 64],
      [225, 64, 237],
      [237, 174, 64],
      ]
], dtype = np.uint8)
io.imshow(color)

ssim_original = structural_similarity(color, color, win_size=3, multichannel = True)
print(ssim_original)
```

```python
median_img = cv2.medianBlur(img, 3)
io.imshow(median_img)
ssim_median = structural_similarity(color, median_img, win_size=3, multichannel = True)
print(ssim_median)
```

```python
gaussian_img = cv2.GaussianBlur(color, (3, 3), 0, cv2.BORDER_DEFAULT)
io.imshow(gaussian_img)
ssim_gaussian = structural_similarity(color, gaussian_img, win_size=3, multichannel = True)
print(ssim_gaussian)
```

```python
bilateral_img = cv2.bilateralFilter(color, 2, 75, 75)
io.imshow(bilateral_img)
ssim_bilateral = structural_similarity(color, bilateral_img, win_size=3, multichannel = True)
print(ssim_bilateral)
```

```python
fig = plt.figure(figsize=(10, 5))

fig.add_subplot(2, 4, 1)
plt.imshow(color)
plt.axis('off')
plt.title("Original Image")
  

fig.add_subplot(2, 4, 2)
plt.imshow(median_img)
plt.axis('off')
plt.title("Median Filtered Image")
  

fig.add_subplot(2, 4, 3)
plt.imshow(gaussian_img)
plt.axis('off')
plt.title("Gaussian Filtered Image")


fig.add_subplot(2, 4, 4)
plt.imshow(bilateral_img)
plt.axis('off')
plt.title("Bilateral Filtered Image")


fig.add_subplot(2, 4, 5)
plt.axis('off')
plt.text(0, 1, ssim_original)


fig.add_subplot(2, 4, 6)
plt.axis('off')
plt.text(0, 1, ssim_median)


fig.add_subplot(2, 4, 7)
plt.axis('off')
plt.text(0, 1, ssim_gaussian)


fig.add_subplot(2, 4, 8)
plt.axis('off')
plt.text(0, 1, ssim_bilateral)
```
